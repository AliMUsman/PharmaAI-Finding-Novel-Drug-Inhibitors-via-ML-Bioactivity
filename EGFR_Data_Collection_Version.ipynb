```python
!pip install chembl_webresource_client # Downloads CHEMBL Client with existing data
from chembl_webresource_client.new_client import new_client
```

```python
import pandas as pd
```

```python
egfr = new_client.target
```

```python
egfr_query = egfr.search("EGFR")
```

```python
targets = pd.DataFrame.from_dict(egfr_query)
targets
```

```python
pref_name = targets.target_chembl_id[3]
```

```python
compound = new_client.activity
```

```python
data = compound.filter(target_chembl_id=pref_name).filter(standard_type="IC50") #Filters out different compounds of the CHEMBL_ID by IC50
```

```python
data_0 = pd.DataFrame.from_dict(data)
data_0
```

```python
data_1 = data_0[data_0.standard_value.notna()] #Filters out values that aren't a number
```

```python
data_1 = data_1[data_0.canonical_smiles.notna()] #
```

```python
data_1
```

```python
data_2 = data_1.drop_duplicates(['canonical_smiles'])
```

```python
values = ['molecule_chembl_id','canonical_smiles', 'standard_value']
```

```python
data_3 = data_2[values]
data_3
```

```python
data_3.to_csv("EGFR_inhibitors_data.csv", index=False)
```
